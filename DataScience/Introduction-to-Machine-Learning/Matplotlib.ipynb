# Line Chart: Stock Price Trend

```python
import matplotlib.pyplot as plt
import numpy as np

# Stock prices for a certain stock over time (e.g., daily closing prices)
days = np.array([1, 2, 3, 4, 5])
prices = np.array([100, 102, 105, 103, 107])

plt.plot(days, prices)
plt.xlabel("Days")
plt.ylabel("Stock Price")
plt.title("Stock Price Trend Over 5 Days")
plt.show()
```

# Line with Markers: Moving Average vs Stock Price

```python
# Stock prices and moving averages
days = np.array([1, 2, 3, 4, 5])
stock_prices = np.array([100, 102, 105, 103, 107])
moving_average = np.array([101, 103, 104, 104, 106])

plt.plot(days, stock_prices, 'o:r', label="Stock Price")
plt.plot(days, moving_average, 's--g', label="Moving Average")
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock Prices vs. Moving Average")
plt.legend()
plt.show()
```

# Bar Plot: Portfolio Allocation

```python
assets = np.array(["Stocks", "Bonds", "Gold", "Crypto"])
allocation = np.array([50, 30, 10, 10])

plt.bar(assets, allocation, color='blue')
plt.xlabel("Asset Classes")
plt.ylabel("Percentage")
plt.title("Portfolio Allocation")
plt.show()
```

# Pie Chart: Asset Allocation in a Portfolio

```python
assets = np.array(["Stocks", "Bonds", "Gold", "Crypto"])
allocation = np.array([50, 30, 10, 10])

plt.pie(allocation, labels=assets, startangle=90)
plt.title("Asset Allocation in Portfolio")
plt.show()
```

# Scatter Plot: Stock Returns vs Volatility

```python
# Stock symbols and their return and volatility (e.g., annualized)
stocks = ["AAPL", "MSFT", "GOOGL", "TSLA", "AMZN"]
returns = np.array([15, 20, 12, 25, 18])  # Returns in %
volatility = np.array([5, 7, 4, 9, 6])   # Volatility in %

plt.scatter(volatility, returns)
for i, stock in enumerate(stocks):
    plt.annotate(stock, (volatility[i], returns[i]))

plt.xlabel("Volatility (%)")
plt.ylabel("Return (%)")
plt.title("Stock Returns vs Volatility")
plt.grid(True)
plt.show()
```

# Histogram: Daily Returns Distribution

```python
# Simulated daily returns (e.g., 250 trading days)
daily_returns = np.random.normal(0, 1, 250)  # Mean=0%, Std=1%

plt.hist(daily_returns, bins=20, color='purple')
plt.xlabel("Daily Return (%)")
plt.ylabel("Frequency")
plt.title("Distribution of Daily Returns")
plt.show()
```

# Multiple Line Plot: Stock Price Comparison

```python
days = np.array([1, 2, 3, 4, 5])
stock_A = np.array([100, 102, 105, 103, 107])
stock_B = np.array([98, 101, 104, 102, 106])

plt.plot(days, stock_A, label="Stock A", color='blue')
plt.plot(days, stock_B, label="Stock B", color='green')
plt.xlabel("Days")
plt.ylabel("Stock Price")
plt.title("Stock Price Comparison: A vs B")
plt.legend()
plt.show()
```

# Grid: Sharpe Ratio vs Risk-Free Rate Sensitivity

```python
risk_free_rates = np.array([1, 2, 3, 4, 5])
sharpe_ratios = np.array([0.5, 0.7, 1.0, 1.2, 1.4])

plt.plot(risk_free_rates, sharpe_ratios, marker='o')
plt.grid(axis='both', linestyle='--', linewidth=0.5)
plt.xlabel("Risk-Free Rate (%)")
plt.ylabel("Sharpe Ratio")
plt.title("Sharpe Ratio Sensitivity to Risk-Free Rate")
plt.show()
```

# Box Plot: Analyze Stock Return Distributions

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Simulate stock returns for three assets
np.random.seed(42)
returns = pd.DataFrame({
    'Stock A': np.random.normal(0.001, 0.02, 1000),
    'Stock B': np.random.normal(0.002, 0.015, 1000),
    'Stock C': np.random.normal(0.0005, 0.03, 1000)
})

# Update the parameter from 'labels' to 'tick_labels'
plt.boxplot([returns['Stock A'], returns['Stock B'], returns['Stock C']], tick_labels=['Stock A', 'Stock B', 'Stock C'])
plt.title("Stock Return Distributions")
plt.ylabel("Returns")
plt.show()
```

# Heatmap: Correlation Matrix of Asset Returns

```python
import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = returns.corr()

# Plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix of Stock Returns")
plt.show()
```

# Area Plot: Portfolio Cumulative Returns

```python
# Simulate stock prices for 3 assets
prices = (1 + returns).cumprod()

# Plot cumulative returns as area plot
plt.fill_between(prices.index, prices['Stock A'], color="skyblue", alpha=0.5, label='Stock A')
plt.fill_between(prices.index, prices['Stock B'], color="orange", alpha=0.5, label='Stock B')
plt.fill_between(prices.index, prices['Stock C'], color="green", alpha=0.5, label='Stock C')
plt.title("Cumulative Returns of Stocks")
plt.ylabel("Cumulative Return")
plt.legend()
plt.show()
```

# Violin Plot: Return Distributions with KDE

```python
sns.violinplot(data=returns)
plt.title("Stock Return Distributions (Violin Plot)")
plt.ylabel("Returns")
plt.show()
```
